# Q1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
soup = BeautifulSoup(page.content)
```

```python
titles = []
for i in soup.find_all(['h1', 'h2','h3','h4','h5','h6'], class_= "mp-h2"):
    titles.append(i.text)
df = pd.DataFrame({'Headers':titles})
df
```

# Q2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame

```python
from bs4 import BeautifulSoup
import requests
page = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
page
```

```python
soup = BeautifulSoup(page.content)

Name = []
for i in soup.find_all('td', class_= "titleColumn"):
    Name.append(i.text.replace("\n"," "))
    


rating = []
for j in soup.find_all('td', class_="ratingColumn imdbRating"):
    rating.append(j.text.replace("\n"," "))
    

year =[]
for k in soup.find_all('span', class_="secondaryInfo"):
    year.append(k.text)
    

import pandas as pd

df = pd.DataFrame({'Name': Name,'Ratings': rating, 'Year of release':year})
df.head(10)
```

# Q4. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape: i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
soup = BeautifulSoup(page.content, 'lxml')
```

```python
table = soup.find('table',class_='table')
headers = []
for i in table.find_all('th'):
    title = i.text
    headers.append(title.replace("\n",""))
df = pd.DataFrame(columns = headers)

for row in table.find_all('tr')[1:]:
    data = row.find_all('td')
    row_data = [td.text.strip().replace("\n","  ")for td in data]
    length = len(df)
    df.loc[length] = row_data
df.head(10)
```

# 6. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.amazon.in/s?k=mobile+phone+under+20000&ref=nb_sb_noss")
page
```

```python
soup = BeautifulSoup(page.content)
name = []
for i in soup.find_all('h2', class_="a-size-mini a-spacing-none a-color-base s-line-clamp-2"):
    name.append(i.text)
    
price = []
for j in soup.find_all('span', class_="a-price-whole"):
    price.append(j.text)
    
imgurl = []
for k in soup.find_all('img', class_="s-image"):
    imgurl.append(k['src'])
    
rating = []
for l in soup.find_all('i', class_="a-icon a-icon-star-small a-star-small-4 aok-align-bottom"):
    rating.append(l.text)
    
import pandas as pd
a ={"Name":name,"Price":price,"ImageURL":imgurl,"Ratings":rating}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
df.head(15)
```

# 8. Write a python program to scrape mentioned details from ‘https://www.dineout.co.in/delhi-restaurants/buffet-special’ : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image url

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
resname =[]
for i in soup.find_all('a', class_="restnt-name ellipsis"):
    resname.append(i.text)

cusine = []
for j in soup.find_all('span', class_="double-line-ellipsis"):
    cusine.append(j.text.split("|")[1])

loc = []
for k in soup.find_all('div', class_="restnt-loc ellipsis"):
    loc.append(k.text)

imgurl = []
for m in soup.find_all('img', class_="no-img"):
    imgurl.append(m['data-src'])

rating=[]
for i in soup.find_all("div", class_="restnt-rating rating-3"):
    rating.append(i.text)
for i in soup.find_all("div", class_="restnt-rating rating-4"):
    rating.append(i.text)

import pandas as pd
df = pd.DataFrame({"Restaurent name":resname,"Cusine":cusine,"Location":loc, "Image URL":imgurl,"Ratings":rating})
df
```

# 9. Write a python program to scrape weather details for last 24 hours from ‘https://en.tutiempo.net/delhi.html?data=last-24- hours’ :i) Hour ii) Temperature iii) Wind iv) Weather condition v) Humidity vi) Pressure

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.tutiempo.net/delhi.html?data=last-24-hours')
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
temp =[]
for i in soup.find_all('td', class_="t Temp"):
    temp.append(i.text)
    
wc = []
for j in soup.find_all('span', class_="thhip ico i0530 u303n"):
    wc.append(j.text)   
for j in soup.find_all('span', class_="thhip ico i0530 u303"):
    wc.append(j.text)

wind = []
for k in soup.find_all('td', class_="wind"):
    wind.append(k.text)
    
humidity = []
for l in soup.find_all('td', class_="hr"):
    humidity.append(l.text)

pressure = []
for m in soup.find_all('td', class_="prob"):
    pressure.append(m.text)
    
Hour = []
for i in soup.find_all ('td',class_='t Temp')[0:24]: 

    if i.previous_sibling.previous_sibling is not None:

        Hour.append(i.previous_sibling.previous_sibling.text)

    else:

        Hour.append(' ')
```

```python
import pandas as pd
a = {"Hour":Hour,"Temperature": temp,"Wind":wind, "Weather Condition":wc, "Humidity":humidity, "Pressure":pressure}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
df.head(24)
```
